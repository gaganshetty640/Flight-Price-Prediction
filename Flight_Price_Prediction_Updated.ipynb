## Total Stops Predictor data processing

```python
data.dropna(inplace=True)
```

```python
Counter(data.Total_Stops)
```

```python
stops = data.Total_Stops.str.split()
stops[1][0]
```

```python
#Converting Total_stops string object into integer object 

for i,stops in enumerate(data.Total_Stops):
    if stops == 'non-stop':
        data.loc[i,'stops'] = 0
    else:
        data.loc[i,'stops'] = int(stops.split()[0])
```

```python
data.stops = data.stops.astype('int')
data.head()
```

## Creating month predictor

```python
for i,date in enumerate(data.Date_of_Journey):
    data.loc[i,'month'] = date.month

data.month = data.month.astype('int')
```

```python
data.head()
```

## Creating weekday predictor

```python
for i,date in enumerate(data.Date_of_Journey):
    data.loc[i,'weekday'] = date.isoweekday()

data.weekday = data.weekday.astype('int')
data.head()
```

```python
data.Dep_Time[1][0:2]
```

```python
for i,time in enumerate(data.Dep_Time):
    data.loc[i,'dep_hour'] = int(time[0:2])

data.dep_hour = data.dep_hour.astype('int')
data.head()
```

# Prediction Model

```python
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=42, test_size=0.3)
model = XGBRegressor(colsample_bytree=0.4603, gamma=0.0468, 
                             learning_rate=0.05, max_depth=3, 
                             min_child_weight=1.7817, n_estimators=2200,
                             reg_alpha=0.4640, reg_lambda=0.8571,
                             subsample=0.5213, silent=1,
                             random_state =42, nthread = -1)
model.fit(X_train, y_train)
y_predict = model.predict(X_test)
```

# Evaluation

```python
import numpy as np
from sklearn.metrics import mean_squared_log_error
1- np.sqrt(mean_squared_log_error(y_test, y_predict))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_predict)
```
